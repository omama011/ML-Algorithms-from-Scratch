```python
#find SW:-similarity within class and SB-similarity between class
n_features=iris.shape[1]
S_W=np.zeros((n_features,n_features))
S_B=np.zeros((n_features,n_features))
for i in class_labels:
    X_c = iris[y == i]
    mean_c=np.mean(X_c,axis=0) 
    S_W += (X_c - mean_c).T.dot((X_c - mean_c))
    n_c = X_c.shape[0]
    mean_diff = np.array(mean_c - mean_feat).reshape(n_features, 1) 
    S_B += n_c * mean_diff.dot(mean_diff.T)
```

```python
V = np.linalg.inv(S_W).dot(S_B) 
```

```python
def eigen_vecs(A, num_iterations=1000, tol=1e-6):  #eigenvals and eigenvectors using power iteration method
    n = A.shape[0]
    
    v = np.random.rand(n)
    v = v / np.linalg.norm(v)
    
    for i in range(num_iterations):
        Av = A @ v
        new_v = Av / np.linalg.norm(Av)
        
        if np.linalg.norm(new_v - v) < tol:
            break
        
        v = new_v
    
    eigenvalue = v @ (A @ v)
    
    return eigenvalue, v

cov_matrix = (V + V.T) / 2
eigenvalue, eigenvector = eigen_vecs(V)

print("Dominant eigenvalue:", eigenvalue)
print("Dominant eigenvector:", eigenvector)
```
